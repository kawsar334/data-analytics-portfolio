```python
import pandas as pd
import numpy as np
import psycopg
from sqlalchemy import create_engine
```

```python
df = pd.read_csv("./pizzas.csv")
df.columns= df.columns.str.lower().str.replace(" ","_")
```

```python
host="localhost"
database="pizza_db"
user="postgres"
password="kawsar"
port=5432
table_name="pizza"
engine = create_engine(f"postgresql://{user}:{password}@{host}:{port}/{database}")
```

```python
df.to_sql(table_name,engine,if_exists="replace",index=False,method="multi",chunksize=1000)
query = "SELECT * FROM pizza LIMIT 5"
df_sql = pd.read_sql(query, engine)
print(df_sql)
print(f"Data inserted into {table_name} table successfully!")
```

```python
order_table= pd.read_csv("orders.csv")
order_table.to_sql("orders",engine,if_exists="replace", index=False, method="multi",chunksize=1000)
query = "SELECT * FROM orders LIMIT 5"
order_sql = pd.read_sql(query, engine)
print(order_sql)
print(f"Data inserted into orders table successfully!")
```
