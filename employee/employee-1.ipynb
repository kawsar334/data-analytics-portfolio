```python
import pandas as pd 
import matplotlib.pyplot as plt 


df = pd.read_csv("employee_info.csv")
df['PER_MONTH_SALARY']=df['PER DAY Salary'] * 30
designation_group = df.groupby("DESIGNATION")['EMP ID'].count()
print(designation_group)
plt.figure(figsize=(16,8))
designation_group.plot(kind="bar")
plt.xticks(rotation=35)
plt.show()
```

```python
company = df.groupby(["Company","DESIGNATION"])['EMP ID'].count()

company.to_csv("company_and_designation.csv")

company_json= pd.read_csv("company_and_designation.csv")
company_json.to_json("company_and_designation.json",orient="records" ,indent=4 )
print(company_json)
plt.figure(figsize=(20, 10))
company.plot(kind="bar")
plt.xticks(rotation=30)
plt.show()

```

```python
plt.figure(figsize=(16, 8))
df['PER DAY Salary'].plot(kind="hist", bins=10)
plt.title("per day salary ")
plt.xticks(rotation=30)
plt.ylabel("FREQUENCY")
plt.show()
```
