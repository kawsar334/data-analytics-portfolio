```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
df = pd.read_csv("sales_and_customer_insights.csv")
print(df.isnull().sum())
```

```python
df.isnull().sum()
```

```python
# df.info()
transaction_counts = df.groupby("Region")['Transaction_ID'].nunique()
plt.figure(figsize=(10,6))
# print(transaction_counts)
transaction_counts.plot(kind="bar")
plt.xticks(rotation=45)
plt.title("Number of Transactions by Region")   
```

```python
df['Peak_Sales_Date']=pd.to_datetime(df['Peak_Sales_Date'])
df['Launch_Date'] = pd.to_datetime(df['Launch_Date'])
df['Peak_Sales_Date']= pd.to_datetime(df['Peak_Sales_Date'])

# print(df['Launch_Date'].dt.daysinmonth)
# print(df['Launch_Date'].dt.day_of_week.value_counts())
```

```python


Purchase_Times = df.groupby('Preferred_Purchase_Times')['Purchase_Frequency'].count()
plt.figure(figsize=(14,10))
print(Purchase_Times)
plt.title("Sales in times") 
Purchase_Times.plot(kind="pie", autopct="%1.1f%%")
plt.show()
```

```python

plt.figure(figsize=(15,10))
info= df['Time_Between_Purchases']
info.plot(kind="hist")
plt.xticks(rotation=30)
plt.yticks(rotation=-10)
plt.show()
```

```python

plt.figure(figsize=(14,10))
Most_Frequent_Category=df.groupby("Most_Frequent_Category")['Purchase_Frequency'].count()
Most_Frequent_Category.plot(kind="pie", autopct="%1.1f%%")
plt.title("purchase Most_Frequent_Category count")
plt.show()
 
```
